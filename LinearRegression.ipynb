```python
def cost(points,m):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i,:14]
        y = points[i,-1]
        for j in range(14):
            # cost caculation wrong
            total_cost += (1/M)*(y - (m*x).sum())**2
        return total_cost

def stepgradient(points,learning_rate,m3):
    m_slope = np.zeros(14)
    M = len(points)
    for i in range(M):
        x = points[i,:14]
        y = points[i,-1]
        for j in range(14):
            m_slope[j] += (-2/M)*(y - (m3*x).sum())*x[j]
    new_m = m3 - learning_rate*m_slope
    return new_m

def gd(points,learning_rate,number_of_iteration):
    m2 = np.zeros(14)
    for i in range(number_of_iteration):
        m2 = stepgradient(points,learning_rate,m2)
        print(i," Cost: ",cost(points,m2))
    return m2

def run ():
    data2 = np.loadtxt("boston_x_y_train.csv",delimiter=',')
    arr = data2[:,0:13]
    df2 = pd.DataFrame(arr)
    df2['C'] = 1
    y = data2[:,13]
    points = df2.values
    scaler = preprocessing.StandardScaler()
    scaler.fit(points)
    points = scaler.transform(points)
    points = np.append(points, y.reshape(-1, 1), axis = 1)
    learning_rate = 0.1
    number_of_iteration = 1000
    m = gd(points,learning_rate,number_of_iteration)
    return m,scaler

def pred():
    t,v = run()
    data = np.loadtxt("boston_x_test.csv",delimiter=',')
    df = pd.DataFrame(data)
    f=df['C'] = 1
    points = df.values
    points = v.transform(points)
    M = len(data)
    y = np.zeros(M)
    for i in range(M):
        x = points[i,:14]
        y[i] += (t*x).sum()
    np.savetxt("predict.csv",y)
```
